## Testing the Ability of the Surprisingly Popular Algorithm to Predict NBA Game Results

### Abstract

Considering the “surprisingly popular” (SP) algorithm to better extract correct answers from a large group of people.<br>
The SP method introduced by Prelec, Seung and McCoy (<a href="https://www.nature.com/articles/nature21054"><i>A solution to the single-question crowd wisdom problem</i>, Nature, Volume 541, Issue 7638, pp. 532-535 (2017), January 2017</a>).

The main claim of the researchers is that algorithms for finding the winning answer for a question are imperfect. Those algorithms usually select the answer according to the majority opinion of the group even if they select the wrong answer. Those algorithms are biased for shallow, lowest common denominator information, at the expense of novel or specialized knowledge that is not widely shared. The researchers claim that even if we consider the responders' confidence, this adjustment does not solve this problem reliably. The researchers suggest a different approach. They claim that under reasonable assumptions about the voter behaviour, choosing the answer which performed better than expected (i.e., is more popular than people predicted), yields the best answer to a single question from the crowd. The researchers use a technique they call the “surprisingly popular” (SP) algorithm to better extract correct answers from a large group of people. The SP method is motivated by the challenge of finding the best answer to a single question, especially in situations where many people in the group could believe in the wrong answer.

### Introduction

In the article the SP method showed very impressive results relative to other methods such as majority voting rule, confidence etc. in range of studies. However, perhaps the most interesting and important potential application of the SP method is to situations requiring genuine prediction where the true answer is not known at the time responders make decisions. None of the studies conducted by the researches were of this type.<br> In this project I will try to test the abilty of the SP method to predict the results of NBA games and to test the preformance of the SP method relative to other methods.

### Data

To test the ability of the SP method to predict NBA games results, I conducted a survey with Google Form and asked the participant to give their predictions for 12 NBA games which took place in January 22nd. I also downloaded FiveThirtyEight predications as experts for comparison. FiveThirtyEight uses three different method to predict the results:
* Elo Rating - a simple measure of strength based on game-by-game result
* CARMELO - a system that forecasts a player‘s future performance
* RAPTOR - Robust Algorithm (using) Player Tracking (and) On/Off Ratings

The RAPTOR method is the latest technique used by FiveThirtyEight to predict NBA games results.<br>
For each matchup the participants were asked to pick the winner of the matchup, give their confidence in the answer and also to predict the precentage of responders they think will pick the same team as they did.
